```python
q_array = {1:q1, 2:q2, 3:q3, 4:q4, 5:q5, 6:q6, 7:q7, 8:q8, 9:q9, 10:q10, 11:q11, 12:q12, 13:q13, 14:q14, 15:q15}
score = 0
money = 0
money_vals = [100, 200, 300, 500, 1000, 2000, 4000, 8000, 16000, 32000, 64000, 125000, 250000, 500000, 1000000]
money_counter = 0
completed = []
```

```python
def ask_question():
    global score
    global money
    global money_counter
    if len(completed) == 15:
        print("You have completed all of the questions")
        print("Your score was " + str(score) + "/15")
        quit()
    
    question_number = random.randint(1, 15)
    if question_number in completed:
        ask_question()
    
    question = q_array[question_number]
    print("***************************************")
    print("This question is worth: £" + str(money_vals[money_counter]))
    q_text = question["q"]
    print("a: " + question["a"])
    print("b: " + question["b"])
    print("c: " + question["c"])
    print("d: " + question["d"])
    print("")
    user_answer = input(q_text)
    
    if user_answer == question["answer"]:
        score = score + 1
        money = money + money_vals[money_counter]
        money_counter = money_counter + 1
        completed.append(question_number)
        print("")
        print("Congratulations! That answer was correct!")
        print("You have £" + str(money) + " in the bank!")
        print("")
        ask_question()
    elif user_answer == "score":
        print("Your current score is: " + str(score) + "/" + str(len(completed)))
        print("You have £" + str(money) + " in the bank!")
        print("")
        ask_question()
    elif user_answer == "end":
        print("Congratulations your score was " + str(score) + "/" + str(len(completed)))
        print("You finished with £" + str(money) + " in the bank!")
        exit()
    else:
        print("")
        print("Sorry that answer was wrong")
        print("Your bank has now reset")
        print("You now have zero in the bank")
        print("The correct answer was " + question["answer"])
        print("")
        money_counter = 0 
        money = 0 
        completed.append(question_number)
        ask_question()
        
```
